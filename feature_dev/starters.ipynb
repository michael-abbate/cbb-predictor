```python
def find_team_starters(team):
    team['num_starts'] = team['starter'].expanding().sum()
    
    team['num_starts'] = team['num_starts'].shift(1).fillna(1)    
    return team

def find_starters_array(team):
    team['starters_array'] = list(team.sort_values(by='num_starts',ascending=False).head()['player'].values).shift(1)
duke_starters = duke.groupby(['team','player'],group_keys=False)[['starter']].apply(find_team_starters)#.rename(columns={'starter':'num_starts'})
```

```python
duke_cleaned = pd.concat([duke, duke_starters],axis=1)
# x.head()
duke_cleaned[duke_cleaned['player']=='K. Filipowski']#[['num_starts']]
```
